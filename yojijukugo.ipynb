```python
df=pd.read_sql_query('SELECT * FROM yojijukugo', con)
```

```python
df
```

```python
yojiKanji = []
yojiKanji2 = []

for i in df['title']:
    temp = []
    
    for k in i:
        yojiKanji.append([k,i])
        temp.append(k)
        
    for j in set(temp):
        yojiKanji2.append([j,i])
```

```python
yojiKanji
```

```python
yojiKanji2
```

```python
dfYojikanji = pd.DataFrame(yojiKanji)
dfYojikanji2 = pd.DataFrame(yojiKanji2)
```

```python
dfKanjiSum = dfYojikanji.groupby([0]).count()
dfKanjiSum2 = dfYojikanji2.groupby([0]).count()
```
