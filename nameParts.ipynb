```python
import urllib.request
import requests
from bs4 import BeautifulSoup
import sys
import os
import re
import numpy as np
import pandas as pd
import time
from datetime import date, timedelta
```

```python
df2 = pd.read_csv('./漢字辞典オンライン.csv')
```

```python
df = pd.read_csv('./name40000_20220429.csv')
```

```python
df = df.rename(columns={'0': 'rank', '1': 'name','2':'sum'})
```

```python
df
```

```python
df['rank'] = df.apply(lambda i: int(re.sub("\\D", "", i['rank'])), axis=1)
```

```python
df['sum']
```

```python
kanjiNum = []

for name,nameSum in zip(df['name'],df['sum']):
    for n in name:
        kanjiNum.append({'name':name,'nameSum':nameSum,'kanji':n})
```

```python
dfKanji = pd.DataFrame(kanjiNum)
```

```python
dfKanji
```

```python
dfKanji = dfKanji.rename(columns={0: 'kanji', 1: 'kanjiSum',2:'lastName'})
```

```python
dfKanji 
```

```python
dfNameParts = pd.merge(dfKanji, df2)[["name","nameSum","kanji","parts"]]
```

```python
dfNameParts.rename(columns={0: 'name', 1: 'nameSum',2:'lastName'})
```
